```python
# ====
# Lab 4.3: 3D Reconstruction with Nerfstudio (Gaussian Splatting)
# Requires: CUDA GPU, photos of an object
# ====

# 1. Install nerfstudio
pip install nerfstudio

# 2. Prepare your photos
# Place 30-50 images in a folder: /data/photos/
# Ensure images are well-lit and cover all angles

# 3. Process the images (estimate camera poses using COLMAP)
ns-process-data images --data /data/photos/ --output-dir /data/my_object

# 4. Train a Gaussian Splatting model (faster than NeRF)
ns-train splatfacto --data /data/my_object

# This will open a web viewer at http://localhost:7007
# You can rotate the camera in real-time and see the 3D reconstruction

# 5. Render a video flythrough
ns-render camera-path --load-config outputs/.../config.yml \
--camera-path-filename camera_path.json --output-path output.mp4

# ====
# Alternative: Use Hunyuan 3D Studio (No Code Required)
# ====
# 1. Go to Hunyuan 3D Studio web interface
# 2. Upload your photos or enter a text prompt
# 3. Download the generated 3D model (GLB/FBX format)
# 4. View in Blender or a web viewer
```
